### KL Div evolution during MAE pretraining

```python
import polars as pl
import matplotlib.pyplot as plt

df = pl.read_csv("...")
```

```python
%matplotlib inline
_ = df.group_by("epoch").agg(
    pl.col(col).sum() for col in df.columns if col not in ["epoch", "step", ]
).sort("epoch")

for col in _.columns:
    if col.startswith("auc_"):
        plt.plot(_["epoch"], _[col], label=col.strip("auc_"))
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.legend()
plt.title("kNN AUROC during MTAE training")
plt.show()
```

```python
%matplotlib inline
_ = df.group_by("epoch").agg(
    pl.col(col).sum() for col in df.columns if col not in ["epoch", "step", ]
).sort("epoch")

for col in _.columns:
    if col.startswith("div_"):
        plt.plot(_["epoch"], _[col], label=col.strip("div_"))
plt.xlabel("Epoch")
plt.ylabel("KL divergence (nats)")
plt.title("KL-div. during MTAE training")
plt.legend()
plt.show()
```
