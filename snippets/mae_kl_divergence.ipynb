```python
import polars as pl
df = pl.read_csv("...")
```

```python
import altair as alt
df.group_by("epoch").agg(
    pl.col(col).sum() for col in df.columns if col not in ["epoch", "step", ]
).unpivot(
    index="epoch",
    on = ["auc_CD", "auc_HYP", "auc_MI", "auc_NORMAL", "auc_STTC"],
    value_name="AUROC",
    variable_name="label",
).plot.line(
    alt.X("epoch"),
    alt.Y("AUROC").scale(domain=(0.7, 0.92)),
    color="label"
)
```

```python
import altair as alt
df.group_by("epoch").agg(
    pl.col(col).sum() for col in df.columns if col not in ["epoch", "step", ]
).unpivot(
    index="epoch",
    on = ["div_CD", "div_HYP", "div_MI", "div_NORMAL", "div_STTC"],
    value_name="div",
    variable_name="label",
).plot.line(
    alt.X("epoch"),
    alt.Y("div").scale(domain=(0.35, 1.5)),
    color="label"
)
```
